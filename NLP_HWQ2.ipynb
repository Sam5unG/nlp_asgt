```python
import pandas as pd
df = pd.read_csv(r'D:\Downloads\IMDB Dataset.csv')

df2= df[:10000][:]
df2.head()
```

```python
rw_words = []
all_words = []
from nltk.tokenize import word_tokenize 
for row in df2.index:
    temp = word_tokenize(df2.review[row])
    rw_words.append((temp, df2.sentiment[row]))
    all_words.extend(temp)    
print(len(rw_words), len(all_words)) 
 

from random import shuffle 
shuffle(rw_words)
```

```python
all_words = [word.lower() for word in all_words]
 
print(all_words[:10], len(all_words))
```

```python
from nltk import FreqDist

all_words_frequency = FreqDist(all_words)
 
print(all_words_frequency)
print(all_words_frequency.most_common(10))
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
 
stopwords_eng = stopwords.words('english')

all_words_wosw = [word for word in all_words if word not in stopwords_eng]

print (all_words_wosw[:10])
```

```python
print(FreqDist(all_words_wosw).most_common(15))
```

```python
import string 
print(string.punctuation)
all_words_woswpunc = [word for word in all_words_wosw if word not in string.punctuation]

print(all_words_woswpunc[:10])
all_words_freq = FreqDist(all_words_woswpunc)
print(all_words_freq.most_common(15))
```

Remove top six words from list as these words don't contribute to sentiment of a review (0,1,4,5,6 - kind of stopwords/punctuation) and (movie, film - generic)

```python
most_common_words = all_words_freq.most_common(4540)
print(most_common_words[7:17])

word_features = [item[0] for item in most_common_words[7:]]
print(word_features[:10])
```

```python
def review_features(review):
    review_words = set(review)
   
    features = {}
    for word in word_features:
        features['contains(%s)' % word] = (word in review_words)
    return features
```

```python
review1 = rw_words[0][0]
review1
```

```python
print(review_features(review1))
```

```python
feature_set = [(review_features(review), sent) for (review, sent) in rw_words]
print(feature_set[0])
```

```python
print(len(feature_set))
 
test_set = feature_set[:2000]
train_set = feature_set[2000:]
 
print(len(train_set))
print(len(test_set))
```

```python
from nltk import NaiveBayesClassifier
 
classifier = NaiveBayesClassifier.train(train_set)
```

```python
from nltk import classify 
 
accuracy = classify.accuracy(classifier, test_set)
print(accuracy) 
```
